### 6.2. Text Generation using models from Huggingface

```python
from transformers import pipeline, set_seed

generator = pipeline('text-generation', model = 'gpt2')
```

```python
set_seed(224)
input_text = "In this research, I find that the most notable differences between Republicans' tweets and Democrats' are"

output = generator(input_text, max_length=80, num_return_sequences=5)
for result in output:
    print(result['generated_text'])
    print("----")
```

```python
output = generator(input_text, max_length=80, num_return_sequences=5, temperature = 1.5)
for result in output:
    print(result['generated_text'])
    print("----")
```
